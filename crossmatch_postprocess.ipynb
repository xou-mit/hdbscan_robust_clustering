```python
# Generate the flags
# Prepare some columns
df['distance'] = 1/(df.parallax-df.zero_point_correction)
df['G'] = df.phot_g_mean_mag

# For the calibrated color factor
# use the expression in Riello et al. (2021)
def f_GBP_GRP(Bp,Rp):
    x = Bp-Rp
    if x < 0.5:
        f = 1.154360+0.033772*x+0.032277*x**2
    elif x >= 0.5 and x < 4.0:
        f = 1.162004+0.011464*x+0.049255*x**2-0.005879*x**3
    elif x >= 4.0:
        f = 1.057572+0.140537*x
    else:
        f = np.nan
    return f

df['color'] = df.phot_bp_mean_mag-df.phot_rp_mean_mag

df['C_cal'] = np.nan
ind_color_tmp = np.where(df.color < 0.5)[0]
df.loc[ind_color_tmp,'C_cal'] = df.loc[ind_color_tmp,'phot_bp_rp_excess_factor'] - (1.154360+0.033772*df.color+0.032277*df.color**2)
ind_color_tmp = np.where((df.color >= 0.5) & (df.color < 4.0))[0]
df.loc[ind_color_tmp,'C_cal'] = df.loc[ind_color_tmp,'phot_bp_rp_excess_factor'] - (1.162004+0.011464*df.color+0.049255*df.color**2-0.005879*df.color**3)
ind_color_tmp = np.where((df.color >= 4.0))[0]
df.loc[ind_color_tmp,'C_cal'] = df.loc[ind_color_tmp,'phot_bp_rp_excess_factor'] - (1.057572+0.140537*df.color)

df['sig_C_cal'] = 0.0059898 + 8.817481*10**(-12)*df.phot_g_mean_mag**(7.618399)
```

```python
# List the cuts
# Keep in mind that the poe > 5 cut and the ruew < 1.4 cut are already applied to the sample
science = (df.parallax_over_error > 5) & (df.distance < 5) 
general = (df.duplicated_source == 0) & (df.visibility_periods_used >= 10)
magnitude = (df.G < 19) #& (df.phot_g_mean_flux_over_error > 2000)
astrometric = (df.astrometric_excess_noise < 2) & (df.ruwe < 1.4)
binary_exclusion = (df.ipd_frac_multi_peak <= 2) & (df.ipd_gof_harmonic_amplitude < np.exp(0.18*(df.G-33)))
photometry = (abs(df.C_cal/df.sig_C_cal) < 3)
rv_dr2 = (abs(df.dr2_radial_velocity) >= 0)
```

```python
ind_qual_gd = np.where(science & astrometric & general & magnitude & photometry)[0]
ind_bin_gd = np.where(binary_exclusion)[0]
ind_tmp = np.where(science & astrometric & general & magnitude & photometry & binary_exclusion)[0]
print(len(ind_qual_gd),len(ind_bin_gd),len(ind_tmp))
```

```python
import uncertainties
from uncertainties import unumpy
import astropy.coordinates as coord
coord.galactocentric_frame_defaults.set('v4.0')
print(coord.Galactocentric())

def celes_to_gc_coord_err(l,b,plx,plx_err):
    '''
    Input units as before
    Output coordinates are in pc
    '''
    rSun = 8122 #pc
    x_shift = -rSun
    y_shift = 0.
    z_shift = 20.8 #pc
    
    plx_u = uncertainties.ufloat(plx, plx_err)
    d_u = 1/plx_u*1000 #pc
    
    x_u = d_u * np.cos(np.radians(b)) * np.cos(np.radians(l)) + x_shift
    y_u = d_u * np.cos(np.radians(b)) * np.sin(np.radians(l)) + y_shift
    z_u = d_u * np.sin(np.radians(b)) + z_shift 
    
    x = unumpy.nominal_values(x_u)
    y = unumpy.nominal_values(y_u)
    z = unumpy.nominal_values(z_u)
    
    x_err = unumpy.std_devs(x_u)
    y_err = unumpy.std_devs(y_u)
    z_err = unumpy.std_devs(z_u)
    
    return x, y, z, x_err, y_err, z_err
```

```python
# Calculate the positions and errors
x_tmp = np.zeros(len(df_vel))
y_tmp = np.zeros(len(df_vel))
z_tmp = np.zeros(len(df_vel))

xerr_tmp = np.zeros(len(df_vel))
yerr_tmp = np.zeros(len(df_vel))
zerr_tmp = np.zeros(len(df_vel))

l_tmp = df_vel['l'].values
b_tmp = df_vel['b'].values
plx_tmp = df_vel['parallax_zpcorr'].values
plxerr_tmp = df_vel['parallax_error'].values

for i in range(len(df_vel)):
# for i in range(20):
    x_tmp[i], y_tmp[i], z_tmp[i], xerr_tmp[i], yerr_tmp[i], zerr_tmp[i] = celes_to_gc_coord_err(l_tmp[i],b_tmp[i],plx_tmp[i],plxerr_tmp[i])
    if i%100000 == 0:
        print(i,'stars finished.')
```

```python
def get_vel_cyl(**kwargs):
    # Create the astropy sample
    samples = coord.SkyCoord(**kwargs)
    vtot = np.sqrt(samples.v_x.to_value(u.km/u.s)**2+samples.v_y.to_value(u.km/u.s)**2+samples.v_z.to_value(u.km/u.s)**2)
    samples.representation_type = 'cylindrical'
    
    return samples.d_rho.to_value(u.km/u.s), samples.d_phi.to_value(u.rad/u.s)*samples.rho.to_value(u.km), samples.d_z.to_value(u.km/u.s)
```

```python
import astropy.units as u
import astropy.coordinates as coord
coord.galactocentric_frame_defaults.set('v4.0')
df_vel['best_vr'], df_vel['best_vphi'], df_vel['best_vz'] = get_vel_cyl(x=x_tmp*u.pc, y=y_tmp*u.pc, z=z_tmp*u.pc, 
                                                                        v_x=df_vel['best_U'].values*u.km/u.second, v_y=df_vel['best_V'].values*u.km/u.second, v_z=df_vel['best_W'].values*u.km/u.second, frame=coord.Galactocentric)
```
