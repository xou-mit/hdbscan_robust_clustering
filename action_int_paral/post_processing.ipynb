```python
# Post processing the calculated result
# Merge them into one calatogue and output as a hdf5 file
# Import the libraries
import numpy as np
import pandas as pd
import glob
```

```python
# Set up the list to hold the result
orb_result_list = []
```

```python
# Group together the orbital parameter results
# Get the file paths
N_cp_list = [5,21,40,46,8,40,33]
folder_list = ['p1','p2','p3','p4','p4_1','p5','p6']
orb_path_list = []
for j in range(len(N_cp_list)):
    N_cp = N_cp_list[j]
    cp_list = np.arange(N_cp)

    for i in range(N_cp):
        orb_file_name = glob.glob(folder_list[j]+'/orb_param_checkpoint_'+str(cp_list[i])+'.csv')[0]
        orb_path_list.append(orb_file_name)
# print(orb_path_list)
N_files = len(orb_path_list)
```

```python
# Import the data and append
for i in range(0,N_files):
    print('Appending file #',i)
    orb_result_list.append(pd.read_csv(orb_path_list[i]))
    print(len(orb_result_list[i]))
```

```python
orb_result_all = pd.concat(orb_result_list,axis=0,ignore_index=True)
```

```python
len(orb_result_all)
```

```python
# Export the data into hdf5 file
orb_result_all.to_hdf('../data/dr3_near_good_cm_orb_param.h5',key='dr3_cm_orb',mode='w')
```
